```python
import os
import numpy as np
import matplotlib.pyplot as plt

from blues.utils import *
plt.style.use('seaborn-talk')
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
def plot_nav_data(dvl, depth, imu, vmax=0.5, dmax=10.0):
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(15, 8))
    ax1.plot(dvl[:, 0], dvl[:, 1], 'r', label='vx')
    ax1.plot(dvl[:, 0], dvl[:, 2], 'g', label='vy')
    ax1.plot(dvl[:, 0], dvl[:, 3], 'b', label='vz')
    ax1.set_ylim(-vmax, vmax)
    ax1.legend(fontsize='x-large', loc='upper right')
    ax1.grid(axis='y')
    ax1.set_ylabel('m/s')

    ax2.plot(dvl[:, 0], dvl[:, 4], 'k', label='altitude')
    ax2.plot(depth[:, 0], depth[:, 1], 'c', label='depth')
    ax2.set_ylim(0, dmax)
    ax2.legend(fontsize='x-large', loc='upper right') 
    ax2.grid(axis='y')
    ax2.set_ylabel('m')
    
    ax3.plot(imu[:, 0], np.rad2deg(imu[:, 7]), 'k', label='roll')
    ax3.plot(imu[:, 0], np.rad2deg(imu[:, 8]), 'c', label='pitch')
#     ax3.plot(imu[:, 0], np.rad2deg(imu[:, 9]), 'orange', label='yaw')
    ax3.legend(fontsize='x-large', loc='upper right') 
    ax3.grid(axis='y')
    ax3.set_ylabel('deg')
```

```python
dataset = '../data/tank/2019-06-18-17-59-42.bag'
dvl, depth, imu = load_nav_data(dataset, progress=False)
plot_nav_data(dvl, depth, imu, 0.5, 2.0)
```
